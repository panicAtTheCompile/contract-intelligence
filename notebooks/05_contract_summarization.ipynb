# Automated Contract Summarization

This notebook generates concise summaries from extracted contract information.

The objective is to reduce the time required to review legal agreements by automatically highlighting key contractual details.

```python
import pandas as pd

df = pd.read_csv("../data/processed/contracts.csv")

df.head()
```

```python
def generate_summary(row):

    summary = f"""
Contract Type: {row['contract_type']}

Borrower: {row['borrower']}

Lender: {row['lender']}

Amount: {row['amount']}

Jurisdiction: {row['jurisdiction']}
"""

    return summary.strip()
```

```python
df["summary"] = df.apply(
    generate_summary,
    axis=1
)

print(df["summary"].iloc[0])
```

```python
def generate_intelligent_summary(row):

    summary = f"""
This {row['contract_type']} involves borrower {row['borrower']} and lender {row['lender']}.

The financed amount is {row['amount']} under the jurisdiction of {row['jurisdiction']}.

The contract contains {len(eval(row['clauses']))} major legal clauses and includes contractual obligations related to payment and compliance requirements.
"""

    return summary.strip()
```

```python
df["intelligent_summary"] = df.apply(
    generate_intelligent_summary,
    axis=1
)

print(
    df["intelligent_summary"].iloc[0]
)
```

```python
summary_df = df[
    [
        "contract_type",
        "borrower",
        "intelligent_summary"
    ]
]

summary_df.head()
```

```python
summary_df.to_csv(
    "../data/processed/contract_summaries.csv",
    index=False
)
```
